##A groupby() operation allows us to examine data on a per category basis.

```python
import pandas as pd
```

```python
df = pd.read_csv('sample.csv')
df
```

Categorical columns are non-continuous

```python
df.groupby(by = 'Category').sum()
```

```python
df.groupby(by = 'Category').mean()
```

```python
df.groupby(by = 'Category').count()
```

```python
df = pd.read_csv('mpg.csv')
df
```

```python
df['model_year'].unique()
```

```python
df['cylinders'].unique()
```

```python
df['cylinders'].value_counts()
```

```python
df.groupby('model_year')
```

```python
df.groupby('model_year').mean()
```

```python
df.groupby('model_year').mean()#['mpg']
```

```python
df.groupby(['model_year', 'cylinders']).mean()
```

```python
df.groupby(['model_year', 'cylinders']).mean().columns
```

```python
df.groupby(['model_year', 'cylinders']).mean().index    #multi-level index
```

```python
df.groupby(['model_year', 'cylinders']).mean()['mpg']
```

```python
df.groupby('model_year').describe().transpose()
```

```python
years_cyl = df.groupby(['model_year', 'cylinders']).mean()
years_cyl
```

```python
years_cyl.index.names
```

```python
years_cyl.index.levels
```

```python
years_cyl.loc[70]
```

```python
years_cyl.loc[[70, 80]]
```

```python
years_cyl.loc[(70,8)]
```

###Cross-Section .xs

```python
years_cyl.xs(key = 70, level = 'model_year')    #single key only
```

```python
years_cyl.loc[[70,80]]
```

```python
#Select only 4 cylinder cars for each year
four_cyl = years_cyl.xs(key = 4, level = 'cylinders')
four_cyl
```

```python
#Only 6 and 8 cylinders
#Filter out the dataframe first, then call the groupby
df[df['cylinders'].isin([6,8])].groupby(['model_year', 'cylinders']).mean()
```

```python
#To swap the levels of model_year and cylinders
years_cyl.swaplevel()
```

```python
years_cyl.sort_index(level = 'model_year', ascending = False)
```

```python
years_cyl.sort_index(level = 'cylinders', ascending = False)
```
