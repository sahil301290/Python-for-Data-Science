##Regular Expressions is also referred to as "regex" or "regexp".

It provides a concise way of matching strings viz. words, characters, or a pattern of characters.

regex is written in a formal language that can be interpreted by regex processor.

###Use cases of regex:

1. Data Mining: Regular expression is the best tool for data mining. It efficiently identifies a text in a heap of text by checking with a pre-defined pattern. Some common scenarios are identifying an email, URL, or phone from a pile of text.
2. Data Validation: Regular expression can perfectly validate data. It can include a wide array of validation processes by defining different sets of patterns. A few examples are validating phone numbers, emails, etc.

Regular Expressions Quick Guide:

regex reference.jpg

Bracket means beginning of the set of characters

Parenthesis ( means start extracting, and ) means stop extracting 

To use regular expression in Python,
"import re"

re.search() can be used to see if string matches a regular expression

```python
with open('Sample.txt') as f:
    hand = f.readlines()
hand
```

```python
for line in hand:
  line = line.rstrip()
  if line.startswith('From'):
    print(line)
```

```python
import re
for line in hand:
  line = line.rstrip()
  if re.search('^From', line):
    print(line)
```

###Wild-Card Characters

The dot character matches any character

If you add the asterik character, the character is "any number of times"

```python
for line in hand:
  if re.search('^A.*:', line):
    print(line)
```

###Match and Extract Data

re.search() returns a T/F depending on whether the string matches the regular expression

If we actually want the matching strings to be extracted, we use re.findall()

```python
a = '01 January 2021 was the start of 2021'
print(re.findall('[0-9]+',a))               #Output of list of strings of all the possible matches
```

```python
  print(re.findall('[AEIOU]+',a)) 
```

###Greedy Matching

The repeat characters * and + push outwards in both directions (greedy) to match the largest possible string

```python
x = 'From: using: AAAA EEE characters'
print(re.findall('^F.+:',x))           #First character in a match is 'F' and the last is ':'
```

### Non Greedy Matching

The ? helps in not being greedy

```python
print(re.findall('^F.+?:',x))
```

###Fine Tuning String Extraction

Separately determine which portion of the match is to be extracted using parenthesis

```python
a = "my email id is qwerty@gmail.com "
re.findall('\S+@\S+',a)           
# \S+    Means at least one non-whitespace character
```

```python
re.findall('is (\S+@\S+)',a)      #Limited by the brackets
```

```python
#Extracting a host name - using find and string slicing [Old Method]
atpos = a.find('@')
atpos
```

```python
sppos = a.find(' ',atpos)
sppos
```

```python
host = a[atpos+1 : sppos]
host
```

```python
#Double Split Pattern
a = "my email id is qwerty@gmail.com"
words = a.split()
email = words[4]
pieces = email.split('@')
print(words)
print(email)
print(pieces)
print(pieces[1])
```

```python
#Extracting using Regular Expression
a = "my email id is qwerty@gmail.com"
print(re.findall('@([^ ]*)',a))     
# @     Look through the string till you find @ sign
# [^ ]  Match non-blank character
# *     Match many of them
```

```python
#Escape character
a = 'Salary of $12000.00 was credited at end of the month'
print(re.findall('\$[0-9.]+',a))
#   \$        A real dollar sign
#   [0-9.]    A digit or period
#   +         At least one or more
```

```python
#Refine the output with specific keywords
a = "From my email id qwerty@gmail.com"
print(re.findall('^From .*@([^ ]*)',a))
```
